# Naive Bayes analysis

## Mathematical Understanding

Starting from the basic Bayes' Theorem:
$$
P(H|E)=\frac{P(E|H)\cdot P(H)}{P(E)}
$$

This says that the posterior probability i.e. the probability that Hypothesis($H$) is true given the Evidence($E$) is updated from the prior probability($P(H)$) using the following formula. Here both $H$ and $E$ are treated as random variables(in the sense that they take their values based on a probability distribution)

In the Naive Bayes classifier, just like in the logistic regression where we used the feature vector $\bold{x}$ as in input into the sigmoid function, we use the $\bold{x}$ feature vector as an evidence to support a hypothesis: whether the incoming data point is in the class $y$:
$$
p(y|\bold{x})=\frac{p(\bold{x}|y)\cdot p(y)}{p(\bold{x})}
$$

Here on, I'll be running with an email filter example since it's easier for me (and probably you) to understand that way. So think of $\bold{x}$ as a list of word counts and associated numerical data to an email so $\bold{x}$ is of the form $\bold{x}=\{x_1, x_2, x_3, x_4, \dots, x_n\}$, and $y$ can take up values such as $\{$ 'Promotions', 'Spam', 'Social'$\}$

Now let's expand the numerator $N$ of the equation we just wrote:
$$
N=p(\bold{x}|y)\cdot p(y)=p(\bold{x}\cap y)\\
=p(\{x_1, x_2,\dots, x_n\}\cap y)=p(x_1\cap x_2\cap \dots x_n\cap y)\\
=p(x_1\cap \{x_2,x_3,\dots, x_n, y\})=p(x_1|\{x_2,x_3, \dots, x_n, y\})\cdot p(\{x_2, x_3, \dots, x_n,y\})\\
=p(x_1|\{x_2,x_3, \dots, x_n, y\})\cdot p(x_2|\{x_3, \dots, x_n,y\})\cdot p(\{x_3, \dots, x_n,y\})\\
=p(x_1|\{x_2,x_3, \dots, x_n, y\})\cdot p(x_2|\{x_3, \dots, x_n, y\})\cdot p(x_3|\{x_4,x_5, \dots, x_n, y\})\cdot \dots p(x_{n-1}|\{x_n, y\})\cdot p(x_n|\{y\})\cdot p(y)
$$
Where we have used the definition of conditional probability repeatedly

Now, the naivete in Naive Bayes analysis comes from the fact that we assume all the variables $x_1, x_2 \dots, x_n$ in the feature vector are independent of each other and only dependent on $y$. So, the last equation becomes:
$$
N=p(x_1|y)\cdot p(x_2|y)\dots p(x_n|y)\cdot p(y)\\
= p(y)\prod_{i=1}^{n}p(x_i|y)
$$

Now, in actuality, we are using this whole mathematical model to make a decision: "Which class do we put the incoming data point in?". Thus, rather than calculating full mathematical value, it allows us a leeway: We don't need to calculate $p(\bold{x})$ since this value is constant no matter what class's probability you are calculating, so it doesn't matter.

Also, to not deal with ever shrinking numbers ($\lim_{n\rightarrow \infty} \prod^{n} x=0$ if $x<1$), we take a log of the numerator. So essentially, when comparing finally we do the comparisons based on:

$$\log(N)= \log(p(y)\prod_{i=1}^{n}p(x_i|y))=\log(p(y))+\sum_{i=1}^{n}\log(p(x_i|y))$$

Now, to make the decision to put the data point in which class(read which value of y), we simply put it in the class(y-value) where the probability(again we use log(N) as the proxy here) is the most:
$$
\hat{y}=\argmax_{k\in {1,\dots, m}} \big(\log(p(y_k)) + \sum_{i=1}^{n} \log(p(x_i|y_k))\big)
$$

In the code below $\log(p(y))$ are called the priors. In regular probability lingo $p(y|x)$ will then be called the posteriors, but in the code we just call $log(N)$ because it's just a proxy for the actual posterior.



Now, we come to the part of how we train this model, which we unravel by asking the question: "How do we find $p(x_i|y)$?" 


We get the answer to this based on the training data, assuming a certain distribution as to how the variables $x_1, x_2,\dots, x_n$ for data points in a particular class $y_k$ are distributed. Here, I have assumed a normal distribution for the variables(which may not be the case, ```x_2``` is certainly a poisson distribution but since calculating pdf for Poisson was causing errors, I had to skip it)

````python
#Actually functioning code
class NaiveBayes():
    # init function not required

    def fit(self, X, y):
        n_samples, n_features = X.shape # type of X and y is numpy.ndarrays
        self._classes = np.unique(y)
        n_classes= len(self._classes)
        
        # init mean, var and prior for x1 since it follows the normal distribution
        self._mean = np.zeros((n_classes, n_features))# Works for gaussian as E(X)=lambda
        self._var = np.zeros((n_classes, n_features))
        self._priors = np.zeros(n_classes)

        for idx, c in enumerate(self._classes):
            X_c = X[y==c] # c=0 false data, c=1 true data
            self._mean[idx, : ] = X_c.mean(axis=0)
            self._var[idx, : ] = X_c.var(axis=0)
            self._priors[idx] = X_c.shape[0] / float(n_samples)

    def predict(self, X):
        y_pred = [self._predict(x) for x in X]
        return np.array(y_pred)

    def _predict(self, x):
        #Basically implementing the Bayes theorem here
        posteriors = []
        for idx,c in enumerate(self._classes):
            prior=np.log(self._priors[idx])
            posterior = np.sum(np.log(self._pdf(idx,x)))
            posterior = posterior + prior
            posteriors.append(posterior)
            #Return the posterior with the highest probability
        return self._classes[np.argmax(posteriors)]
    
    def _pdf(self, class_idx, x):
        mean = self._mean[class_idx]
        var = self._var[class_idx]
        numerator = np.exp(-((x - mean) ** 2) / (2 * var))
        denominator = np.sqrt(2 * np.pi * var)
        return numerator / denominator
    
    
````
